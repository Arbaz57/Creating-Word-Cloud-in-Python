# What is Word Cloud?

A word cloud is an image of several words. The size of the words represent the frequency of that word in our data.
##### Word Cloud is also refered as 'Tag Cloud'.

### Where is Word Cloud used?

1) Top Hashtags on Social Media(Instagram, Twitter): Throughout the world, social media is trending for the latest updates, so from that we can get the most used Hashtags that people use in their posts.

2) Hot Topics In Media: Analyzing the news articles, we can find the keywords in the headlines and extract the top n demanding topics and to get the desired result i.e the top n trending media topics.

3)Search Term in an E-commerce: In an e-commerce shopping website, the owner can make the word cloud of the shopping items that have been searched the most. So, he can get the idea about which shopping is in great demand during a specific period.

#### In this project we will be creating Word-Cloud from speech "Mann Ki baat 2.0" given from our PM Narendara Modi.

#### Refrence for dataset: https://www.pmindia.gov.in/en/tag/pmspeech/

```python
import glob # We will use glob library to create a list of our file name
read_files = glob.glob("*.txt") # Create a list of all our file names
with open("result.txt", "wb") as outfile: # Opening and creating result file in write mode
    for f in read_files: # Iterate through the list
        with open(f, "rb") as infile: # Open each file in read mode
            outfile.write(infile.read()) # Read the data and write it in our result file
```

```python
import matplotlib.pyplot as plt # Importing to show our image
from wordcloud import WordCloud, STOPWORDS # From word cloud immporting WordCloud and Stopwords
file = open('result.txt', encoding="utf8") # Opening our file in encoding
data = file.read().replace('\n', '') # Converting Our file in a single string
wordcloud = WordCloud(stopwords=STOPWORDS,background_color = 'white',width=1800,height=1000).generate(data)
# Genrating word cloud without any masking
plt.figure(figsize=[18,10])
plt.imshow(wordcloud) #To display image
plt.axis('off')
plt.show()
```

### Updating Stopwords

```python
stopwords = set(STOPWORDS)
stopwords.update(["Mann", "Ki", "Baat", "Prime", "Minister"])
```

```python
wordcloud = WordCloud(stopwords=STOPWORDS,background_color = 'white',width=1800,height=1000).generate(data)
# Genrating word cloud without any masking
plt.figure(figsize=[18,10])
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

## Creating word cloud on mask image

```python
from PIL import Image # From pillow library importing image
import numpy as np
cloud_mask = np.array(Image.open('Cloud.png'))  # creating an array of our image
wordcloud = WordCloud(stopwords=STOPWORDS,background_color = 'white',mask= cloud_mask, 
                      contour_width=3,contour_color='blue').generate(data)
#Contour will give out line to our mask
plt.figure(figsize=[18,10])
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

### Saving the generated Word Cloud

```python
wordcloud.to_file("ourcloud.png")
```
